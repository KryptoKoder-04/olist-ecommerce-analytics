### RFM Customer Segmentation

R (Recency) → Days since last purchase

F (Frequency) → Total orders per customer

M (Monetary) → Total revenue per customer

```python
# Make sure timestamp is datetime
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])

# Reference date (max date in dataset)
reference_date = df['order_purchase_timestamp'].max()
```

```python
rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp': lambda x: (reference_date - x.max()).days,  # Recency
    'order_id': 'nunique',  # Frequency
    'revenue': 'sum'  # Monetary
}).reset_index()

rfm.columns = ['customer_unique_id', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```


### Recency Scoring 
Lower recency = better customer ;
So reverse scoring.

```python
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
```

```python
rfm['R_score']
```

### Frequency Scoring
Higher frequency = better

```python
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
```

### Monetary Scoring
Higher spending = better

```python
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

### Create RFM Score

```python
rfm['RFM_Score'] = (
    rfm['R_score'].astype(str) +
    rfm['F_score'].astype(str) +
    rfm['M_score'].astype(str)
)
```

```python
rfm['RFM_Score'].head()
```

```python
def segment_customer(row):
    if row['R_score'] == 5 and row['F_score'] == 5 and row['M_score'] == 5:
        return 'Champions'
    elif row['R_score'] >= 4 and row['F_score'] >= 4:
        return 'Loyal Customers'
    elif row['R_score'] <= 2:
        return 'At Risk'
    else:
        return 'Regular'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
rfm['Segment'].head()
rfm['Segment'].value_counts()
```

## STEP 5 – CUSTOMER & PROFIT INTELLIGENCE FRAMEWORK


### 1. Revenue Contribution by Customer Segment
- Merges RFM segments to orders
- Calculates revenue, profit, orders per segment
- Adds revenue percentage per segment
- **Purpose**: See which customer types (Champions, Loyal, At Risk…) generate most money
- **Main value**: Guides where to focus retention & marketing budget

### 2. Pareto Analysis (80/20 Rule)
- Sorts customers by descending revenue
- Calculates cumulative revenue percentage
- **Purpose**: Shows how concentrated revenue is among few customers
- **Main value**: Reveals business risk if top customers leave + supports VIP/loyalty strategy

### 3. Cohort Retention Analysis
- Assigns cohort = first purchase month
- Calculates months since first purchase (cohort index)
- **Purpose**: Tracks how retention drops over time after first buy
- **Main value**: Identifies when customers disappear → timing for re-engagement campaigns

### 4. Profitability Deep Dive (by category)
- Revenue + profit per product category
- Calculates profit margin % per category
- **Purpose**: Shows which categories are actually profitable (not just high revenue)
- **Main value**: Find loss-making products → raise prices, stop selling, renegotiate suppliers

### 5. Customer Lifetime Value (CLV) – basic version
- Total revenue + number of orders per customer
- Calculates average order value (AOV)
- **Purpose**: Estimates historical value per customer
- **Main value**: Sets maximum acquisition cost + foundation for future CLV prediction

### 6. Churn Risk Identification
- Flags customers with Recency > 180 days as High risk
- **Purpose**: Quick way to spot customers likely already gone or about to leave
- **Main value**: Creates immediate win-back list for emails, offers, discounts

### One-line summary of Step 5 overall
- Moves from basic KPIs → actionable intelligence about valuable customers, profit leaks, retention timing and churn danger.

### Cohort Retention Analysis (Advanced)
