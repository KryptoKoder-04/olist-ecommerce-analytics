#  COMPLETE INDUSTRY-GRADE INTERVIEW STRUCTURE


# ✅ SECTION 1: PROJECT FOUNDATION & BUSINESS UNDERSTANDING

## 1️⃣ What Was the Business Problem?

### How To Answer (Structure):

**1. Business Context**

* E-commerce / Retail / SaaS / FMCG?
* B2B or B2C?
* Revenue stagnation? Profit decline? Customer churn?

**2. Problem Statement Examples**

* Revenue growth slowing
* Profit margin declining
* High logistics cost
* Low retention
* Regional imbalance
* Over-dependency on few customers

**3. Strategic Objective**

* Increase profit margin by X%
* Improve retention by X%
* Reduce freight cost ratio
* Improve category-level profitability

---

## 2️⃣ What Were Your Key KPIs?

### Core KPIs (Must Know Formula)

| KPI            | Formula                     | Why Important   |
| -------------- | --------------------------- | --------------- |
| Total Revenue  | SUM(Sales)                  | Business scale  |
| Total Profit   | SUM(Profit)                 | Sustainability  |
| Profit Margin  | Profit / Revenue            | Efficiency      |
| AOV            | Revenue / Orders            | Basket strength |
| Retention Rate | Returning Customers / Total | Loyalty         |
| Churn Rate     | Lost Customers / Total      | Risk            |
| Freight Ratio  | Freight / Revenue           | Cost leakage    |

👉 Interviewer expects:

* Metric definition
* Calculation logic
* Business meaning

---



# ✅ SECTION 2: DESCRIPTIVE ANALYTICS (WHAT HAPPENED?)


## 🔹 Revenue & Growth


### Questions:

1. Total revenue & profit?
2. Month-over-Month (MoM) growth?
3. Seasonality?
4. Revenue trend?
5. AOV trend?

### What You Should Explain:

* Monthly aggregation
* Using GROUP BY month
* MoM growth = (Current - Previous)/Previous
* LAG() window function
* Seasonal spikes (festive months)

### Business Insight Example:

* Revenue peaks during Q4
* Q2 shows slow movement
* AOV declining → possible discounting

---

## 🔹 Category Analysis

### Key Interview Questions:

* Highest revenue category?
* Highest margin category?
* Any revenue-rich but profit-poor categories?

### Deep Thinking:

High revenue ≠ High profit
Low margin = risk

### Business Insight:

* Electronics → High revenue but low margin
* Furniture → High freight cost
* Strategy → Price optimization / bundle products

---


## 🔹 Regional Performance



### Evaluate:

* Revenue by state
* Margin by state
* Freight cost ratio by state

### Advanced Angle:

* Underperforming states due to:

  * Logistics cost
  * Low order volume
  * Discount-heavy sales

### Strategic Question:

“Should we expand in that state?”

Answer depends on:

* Profit margin
* Market potential
* Cost structure

---

# ✅ SECTION 3: DIAGNOSTIC ANALYTICS (WHY HAPPENED?)


## Q: Revenue is increasing but profit decreasing. Why?

Possible Reasons:

1. High discounting
2. Freight increase
3. High return rate
4. Low-margin product mix
5. Acquisition cost increasing

You must explain:

* Data analysis method
* Metric comparison
* Root cause identification

---

## Q: Why are margins low?

Check:

* Freight ratio
* Category-level margin
* Discount %
* Return orders

---

## Q: Why customers churning?

Check:

* Recency > 90 days?
* Decreasing frequency?
* Low engagement?

# ✅ SECTION 4: CUSTOMER ANALYTICS (ADVANCED LEVEL)

## 🔵 RFM Analysis


### Recency:

Days since last purchase

### Frequency:

Number of purchases

### Monetary:

Total spending

---



### Segments:

* Champions
* Loyal
* Potential
* At Risk
* Lost

---


### Important Interview Points:

* % customers in each segment
* Revenue contribution of top segment
* Strategy for At Risk

Example Strategy:

* Offer targeted discount
* Loyalty reward
* Email reactivation

---


## 🔵 Pareto (80/20 Rule)

### What to Explain:

* Sort customers by revenue
* Calculate cumulative %
* Identify top 20%

### Business Risk:

High dependency on few customers = vulnerability

---



## 🔵 Cohort Analysis

### Explain Structure:

* Group customers by first purchase month
* Track retention over time

### Metrics:

Month 1 retention
Month 2 retention
Month 3 retention

### Insight:

Most churn happens after month 2
